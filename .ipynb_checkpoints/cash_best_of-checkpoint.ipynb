```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
tickers = ['SHV', 'FXE', 'FXY', 'FXF', 'FXB', 'FXC', 'FXA', 'CYB', 'GLD']

# Download historical data for the ETFs
data = yf.download(tickers, start='2010-01-01')['Adj Close']

# Plot the ETFs in a logarithmic scale
fig, ax = plt.subplots(figsize=(10, 6))
data.plot(ax=ax, logy=True)

# Add a legend
plt.title("Cash competition")
plt.legend()
# Show the plot
plt.show()
```

```python
returns = data.pct_change(252)
fig, ax = plt.subplots(figsize=(10, 6))
returns.plot(ax=ax)
# Add a legend
plt.title("12 months rolling returns")
plt.legend()
# Show the plot
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Define the list of ETFs to track
tickers = ['SHV', 'FXE', 'FXY', 'FXF', 'FXB', 'CYB', 'GLD']

# Download historical data for the ETFs
data = yf.download(tickers, start='2010-01-01')['Adj Close']

# Calculate the 12-month rolling returns for each ETF
returns = data.pct_change(252)

# Define the starting capital
capital = 10000

# Initialize variables for tracking the current number 1 ETF and the equity curve
current_etf = ''
equity_curve = [capital]

# Loop over each day of the returns data
for date in returns.index:
    # Calculate the current rankings of the ETFs based on their 12-month rolling returns
    rankings = returns.loc[date].rank(ascending=False)
    if rankings[rankings == 1].empty:
        continue
    print(rankings[rankings == 1].name)
    print(rankings[rankings == 1].index[0])
```
