```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
tickers = ['SHV', 'FXF', 'GLD']

# Download historical data for the ETFs
data = yf.download(tickers, start='2008-01-01')['Adj Close']

# Plot the ETFs in a logarithmic scale
fig, ax = plt.subplots(figsize=(10, 6))
data.plot(ax=ax, logy=True)

# Add a legend
plt.title("Cash competition")
plt.legend()
# Show the plot
plt.show()
```

```python
returns = data.pct_change(252)
fig, ax = plt.subplots(figsize=(10, 6))
returns.plot(ax=ax)
# Add a legend
plt.title("12 months rolling returns")
plt.legend()
# Show the plot
plt.show()
```
