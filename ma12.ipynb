```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# Step 2: Download historical data for S&P 500 index
data = yf.download('^GSPC')

# Step 3: Resample the data to monthly frequency
monthly_data = data.resample('M').last()

# Step 4: Calculate the 12-month simple moving average
monthly_data['sma12'] = monthly_data['Close'].rolling(window=12).mean()

# Step 5: Generate trading signals based on the crossover strategy
monthly_data['signal'] = 0
monthly_data['signal'][12:] = np.where(monthly_data['Close'][12:] > monthly_data['sma12'][12:], 1, 0)
monthly_data['positions'] = monthly_data['signal'].diff()

# Step 6: Backtest the strategy and calculate returns
monthly_data['strategy_returns'] = monthly_data['Close'].pct_change() * monthly_data['signal'].shift(1)
monthly_data['market_returns'] = monthly_data['Close'].pct_change()

# Calculate the cumulative returns
monthly_data['strategy_cumulative_returns'] = (1 + monthly_data['strategy_returns']).cumprod()
monthly_data['market_cumulative_returns'] = (1 + monthly_data['market_returns']).cumprod()

# Step 7: Plot the results
plt.figure(figsize=(10, 5))
plt.plot(monthly_data['strategy_cumulative_returns'], label='Strategy')
plt.plot(monthly_data['market_cumulative_returns'], label='Market')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Cumulative Returns')
plt.title('S&P 500 Momentum Strategy Backtest')
plt.show()
```
